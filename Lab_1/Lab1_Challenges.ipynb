## Challenge 4

The following code uses a for loop to calculate the factorial of a number.

```python
factorial = 1
num = 5

# for loop to calculate factorial
for j in range(1, num+1):
    factorial *= j
    print(f"Factorial: {factorial}")
```

## Challenge 8 

The code below produces two pandas dataframes. Subsequently, we manipulate the dataframe and calculate some basic statistics including:

- Display the first 5 rows of the DataFrame 'sales_data'

- Use loc to filter and display rows where the "Quantity" is greater than 10 and the "Region" is "North".

- Use iloc to display the values in the first row and the first three columns.

- Calculate the total sales for each region and display the result.

- Display the last 3 rows of the DataFrame 'movie_ratings'.

- Calculate and display the average rating for each genre.

```python
import pandas as pd

sales_data_dict = {
    'Date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04'],
    'Region': ['North', 'South', 'East', 'West'],
    'Product': ['A', 'B', 'A', 'C'],
    'Quantity': [15, 8, 12, 10],
    'Revenue': [1500, 1200, 1800, 900]
}

sales_data = pd.DataFrame(sales_data_dict)


movie_ratings_dict = {
    'Title': ['Movie1', 'Movie2', 'Movie3', 'Movie4'],
    'Genre': ['Action', 'Drama', 'Comedy', 'Thriller'],
    'Year': [2015, 2010, 2018, 2011],
    'Rating': [8.5, 7.8, 9.2, 8.0]
}

movie_ratings = pd.DataFrame(movie_ratings_dict)

print(sales_data)
print(movie_ratings)
```

```python
# Display first 5 rows
print(sales_data.head())

# Filter rows
sales_filtered = sales_data.loc[(sales_data['Quantity'] > 10) & (sales_data['Region'] == "North")]

# Display values
print(sales_data.iloc[0])
print(sales_data.iloc[:, :3])

# Last three rows
print(movie_ratings.tail(3))

# Calculate total sales per each region 
sales_by_region = sales_data.groupby('Region')['Revenue'].sum()
print(sales_by_region)

# Calculate and display average rating by genre
average_ratings_by_genre = movie_ratings.groupby('Genre')['Rating'].mean()
print(average_ratings_by_genre)
```

## Challenge 9: Geopandas and Rasterio
The following code uses libraries for manipulation of spatial data geopandas and rasterio to make simple visualizations of a dataset on road accidents. First, we create an interactive map which visualises locations of accidents and then speed limits using a simple chloropleth map.

Next, we work with elevation data. First, we visualize elevation in a map and subsequently, we create a histogram.

```python
import geopandas as gpd
import rasterio
from rasterio.plot import show

shapefile_path = 'Lab1_Data/Road_Safety_Accidents.shp'
gdf = gpd.read_file(shapefile_path)
gdf.explore()
```

```python
gdf.explore('SPEED_LIMI', cmap='Blues')
```

```python
# import a raster data showing elevations
raster_path = 'Lab1_Data/elev.tif'
raster = rasterio.open(raster_path)
show(raster)
```

```python
# Create histogram of elevations
from rasterio.plot import show_hist
show_hist(raster, bins=50, stacked=False, alpha=0.3, 
          histtype='stepfilled', title="Histogram")
```
